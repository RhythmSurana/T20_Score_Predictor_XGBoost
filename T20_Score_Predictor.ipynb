```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
from glob import glob
```

```python
base_path = os.path.join('cricket_pred', 't20s')
filenames = glob(os.path.join(base_path, '*.yaml')) # Change to *.json if needed

all_dfs = []
counter = 1

print(f"Found {len(filenames)} files. Starting processing...")

# 2. Process files with error handling
for file_path in tqdm(filenames[-1000:-900]):
    try:
        with open(file_path, 'r') as f:
            data = safe_load(f)
            
            # Validation: Fixes the NotImplementedError
            if data is None or not isinstance(data, (dict, list)):
                continue
                
            # Flatten the data
            df = pd.json_normalize(data)
            
            # Add metadata
            df['match_id'] = counter
            df['source_file'] = os.path.basename(file_path)
            
            all_dfs.append(df)
            counter += 1
            
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

# 3. Combine everything
if all_dfs:
    final_df = pd.concat(all_dfs, ignore_index=True)
    print("Success! Combined DataFrame shape:", final_df.shape)
else:
    print("No valid data found.")

# Display result
final_df.tail()
```

```python
matches = pickle.load(open('dataset_level1.pkl','rb'))
matches.iloc[0]['innings'][0]['1st innings']['deliveries']
```

```python
count = 1
all_deliveries = []   # 🔥 store dfs here instead of appending

for index, row in matches.iterrows():
    
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count += 1
        continue
    
    count += 1
    
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    
    try:
        deliveries = row['innings'][0]['1st innings']['deliveries']
        team_name = row['innings'][0]['1st innings']['team']
        
        for ball in deliveries:
            for key in ball.keys():
                match_id.append(count)
                batting_team.append(team_name)
                teams.append(row['info.teams'])
                ball_of_match.append(key)
                batsman.append(ball[key]['batsman'])
                bowler.append(ball[key]['bowler'])
                runs.append(ball[key]['runs']['total'])
                city.append(row.get('info.city', 'Unknown'))
                venue.append(row.get('info.venue', 'Unknown'))
                
                # safer wicket handling
                player_of_dismissed.append(
                    ball[key].get('wicket', {}).get('player_out', '0')
                )
    
    except Exception:
        continue
    
    loop_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue
    })
    
    all_deliveries.append(loop_df)

# 🔥 ONE concat at end
delivery_df = pd.concat(all_deliveries, ignore_index=True)

delivery_df.head()
```

```python
output['runs_x'] = pd.to_numeric(output['runs_x'])

output['current_score'] = (
    output
    .groupby('match_id')['runs_x']
    .cumsum()
)
```

```python
output['over'] = output['ball'].apply(lambda x:str(x).split(".")[0])
output['ball_no'] = output['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
output['balls_bowled'] = (output['over'].astype('int')*6) + output['ball_no'].astype('int')
```

```python
output['crr'] = round((output['current_score']*6)/output['balls_bowled'],2)
```

```python
output['player_dismissed'] = output['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
output['player_dismissed'] = output['player_dismissed'].astype('int')
output['player_dismissed'] = pd.to_numeric(output['player_dismissed'])
output['player_dismissed'] = output.groupby('match_id')['player_dismissed'].cumsum()
output['wickets_left'] = 10 - output['player_dismissed']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
